```python
import matplotlib.pyplot as plt
%matplotlib inline
import datetime
import numpy as np
import pandas as pd
```

```python
term_topic =  pd.read_csv("c:/cygwin64/home/wanghuaq/term_topic.csv", skipinitialspace=True)
```

```python
term_topic
```

```python
#term_topic1 = term_topic.groupby('topic')
term_topic['rank'] = term_topic.groupby('topic')['weight'].rank(ascending=False)
term_topic
```

```python
term_topic_10 = term_topic[term_topic['rank']<=3]
term_topic_10
```

```python
terms = term_topic_10['term']
```

```python

vocab = list(set(terms.values.tolist()))
```

```python
vocindexdict = dict(zip(vocab, range(0, len(vocab))))
indexvocdict = dict(zip( range(0, len(vocab)), vocab))
```

```python
term_topic_10['index'] = [vocindexdict[x] for x in terms.values]
```

```python
# -*-coding: utf-8 -*- 
numtopics = 30
matplotlib.rcParams['svg.fonttype'] = 'none'
y = term_topic_10['index'].values
x = term_topic_10['topic'].values
s = term_topic_10['weight'].values * 40000
y_ticks = [indexvocdict[i] for i in y ]
x_ticks = ["Topic"+str(i) for i in x]

fig = plt.figure(figsize=(10,12))
#plt.rcParams['font.sans-serif'] = ['Microsoft YaHei']
#plt.rcParams['axes.unicode_minus'] = False

ax = fig.gca()
ax.xaxis.tick_top()
plt.yticks(y, y_ticks)
plt.xticks(x, x_ticks, rotation=60)
plt.grid()

plt.tight_layout()
plt.scatter(x, y, s=s, marker='o', c=x)
#term_topic_10.plot(ax=ax,kind='scatter', x='topic', y='index', figsize=(9,12))
plt.show()
```
