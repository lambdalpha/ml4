# Restricted Boltzmann Machines
### 1. Energy based models(EBM)
- Markov random field(Undirected graphical Models)
- Energy based models
Probability $$p(x) =\frac{e^{-E(x)}}{Z}$$
partition function $$Z = \sum_x e^{-E(x)}$$
log-Likelihood: $$ L(\theta, D) = \frac{1}{N}\sum_{x^{(i)}\in D}\log p(x^{(i)})$$
Negative log-Likelihood: $$l(\theta,D) = - L(\theta,D)  $$

### 2. EBMs with hidden units
- Observed part
$$p(x) = \sum_h p(x, h) = \sum_h {\frac{e^{-E(x,h)}}{Z}}$$

- Free Energe
$$ F(x) = -\log \sum_h e^{-E(x,h)} $$

Then we can write
$$ p(x) = \frac{e^{-F(x)}}{Z} with Z = \sum_x{e^{-F(x)}}$$

- The data negative log-likelihood gradient
$$ - \frac{\partial \log p(x)} {\partial \theta} = \frac{\partial F(x)}{\partial \theta} - \sum_\widetilde{x} p(\widetilde{x})\frac{\partial F(\widetilde{x})}{\partial \theta} $$

The first term increases the probability of training data(by reducing the corresponding free energy), while the second term decreases the probability of samples generated by the model.
The gradient is difficult to be determined analytically, as it includes $E_P[\frac{\partial F(x)}{\partial \theta}]$. This require to compute the expectation over all possible configurations of the input x.


A first step to make the computation trackable is to estimate the expectation using a fixed number of model samples. Samples used to estimate the negative phase gradient are reffered to as negative particles, with are donated as **N**.

- the gradient
$$ - \frac{\partial \log p(x)} {\partial \theta} \approx \frac{\partial F(x)}{\partial \theta} - \frac{1}{|N|}\sum_{\widetilde{x} \in N}\frac{\partial F(\widetilde{x})}{\partial \theta} $$

- How to sample N <br>
  **MCMC**



### 3. Restricted Boltzmann Machines(RBM)
- Boltzmann Machines are a particular form of log-linear Markov Random Field(MRF), in which the energy function is linear in its free parameters. <br>
- Using unobserved varialbes to represent complicated distributions.
- By using more hidden variables, we can increase the model capacity of BM.
- RBMs restrict BMs to those without visible-visible and hidden-hidden connections.
- Energy function of RBM
$$ E(v,h) = -b^Tv - c^Th -h^TWv$$
$W$ represents the weights connecting hidden and visible units, and b,c are the offsets of the visble and hidden layers respectively.
- Free energy formular
$$ F(x) = -b^Tv - \sum_i\log \sum_{h_i}e^{h_i(c_i+W_iv)}$$
- visible units and hidden units are conditionally independent given one-another:
$$p(h|v) = \prod_i p(h_i|v_i)$$
$$p(v|h) = \prod_i p(v_i|h_i)$$
